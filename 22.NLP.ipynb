### 워드클라우드

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
```

```python
text = open('../06_alice.txt').read()
print(text)
```

```python
#png파일을 열어서 np.array로 변환
alice_mask = np.array(Image.open('../06_alice_mask.png'))
alice_mask
```

```python
stopwords = set(STOPWORDS)
stopwords
```

```python
stopwords.add('said')
```

```python
import matplotlib.pyplot as plt
import set_matlotlib_hangul
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
```

```python
wc = WordCloud(
    background_color='white',
    max_words=2000,
    mask=alice_mask,
    stopwords=stopwords
)

wc = wc.generate(text)
```

```python
#단어들의 발생 빈도(상대적 수치)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

## 육아휴직관련 법안
- KoNLPy 는 대한민국 법령을 가지고 있다

```python
import nltk
from konlpy.corpus import kobill

doc_ko = kobill.open('1809890.txt').read()
```

```python
print(doc_ko)
```

```python
#Okt 엔진으로 명사분석
from konlpy.tag import Okt

t = Okt()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
#nltk 엔진으로 토큰(빈도수 포함)분석

ko = nltk.Text(tokens_ko, name= '육아휴직법')
ko
```

```python
len(ko.tokens)
```

```python
#중복된 명사 제거한 단어수
len(set(ko.tokens))
```

```python
ko.vocab()
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
#그래프를 보니 의미없는 단어가 많아 stopwords에 추가
#한글은 영어처럼 깔끔하게 분리되지 않아서 보통은 손으로 잡아서 제거
```

```python
data = ko.vocab().most_common(150)

wordcloud = WordCloud(
    font_path='../Fonts/malgun.ttf',
    relative_scaling=0.2,
    background_color='white'
).generate_from_frequencies(dict(data))

plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
