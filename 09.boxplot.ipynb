- IQR에서 +-1.5배에서 벗어난 지점에 있다면 그 데이터는 아웃라이어다
- 

```python
samples = [1,7,9,16,36,38,45,46,48,51,100,101]
```

```python
samples
```

```python
temp_y = [1]*len(samples)
```

```python
temp_y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,4))
plt.scatter(samples,temp_y)
plt.grid()
plt.show()
```

```python
import numpy as np
#중앙값
np.median(samples)
```

```python
#1사분위수
np.percentile(samples,25)
```

```python
#3사분위수
np.percentile(samples,75)
```

```python
q1 = np.percentile(samples,25)
q2 = np.median(samples)
q3 = np.percentile(samples,75)
iqr = q3 - q1
upper_fence = q3 + 1.5*iqr
lower_fence = q1 - 1.5*iqr
```

```python
plt.figure(figsize=(12,4))
plt.scatter(samples,temp_y)
plt.axvline(x=q1, color='k')
plt.axvline(x=q2, color='red')
plt.axvline(x=q3, color='k')
plt.axvline(x=upper_fence, color='k', ls= 'dashed')
plt.axvline(x=lower_fence, color='k', ls= 'dashed')
plt.show()
```
