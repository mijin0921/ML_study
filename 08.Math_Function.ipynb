x축 방향이동

```python

x = np.linspace(-3,2 ,100)
y = 3* x**2 + 2
y2 = 3* (x+1)**2 + 2
```

```python
import matplotlib as mpl

mpl.style.use('seaborn-whitegrid')

plt.figure(figsize=(12,8))
plt.plot(x,y, lw=2,ls='dashed',label='$y=3x^2 +2$')
plt.plot(x,y2, label='$y=3(x+1)^2 +2$')
plt.legend()
plt.grid()
plt.xlabel('$x$',fontsize=20)
plt.ylabel('$y$',fontsize=20)
plt.show();
```

## 지수함수

```python
x = np.linspace(-2,2,100)

a11,a12,a13 = 2,3,4
y11,y12,y13 = a11**x, a12**x, a13**x

a21,a22,a23 = 1/2, 1/3, 1/4
y21,y22,y23 = a21**x, a22**x, a23**x
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x, y11,color='k',label = '$2^x$')
ax[0].plot(x, y12,'--',color='k',label = '$3^x$')
ax[0].plot(x, y13,':',color='k',label = '$4^x$')

ax[0].legend(fontsize=15)


ax[1].plot(x, y21,color='k',label = '$(1/2)^x$')
ax[1].plot(x, y22,'--',color='k',label = '$(1/3)^x$')
ax[1].plot(x, y23,':',color='k',label = '$(1/4)^x$')

ax[1].legend(fontsize=15)

plt.show()
```

지수에 - 가 있으면 그래프가 대칭이 된다.

$$2^{-x}$$

지수 증가라는 단어의 위력

```python
x = np.linspace(0,10)

plt.figure(figsize=(6,6))
plt.plot(x,x**2,'--',color='k',label='$x^2$')
plt.plot(x,2**x,color='k',label='$2^x$')
plt.legend()
```

매우 특이한 지수식

$$\left(1 + \frac{1}{x}\right)^x$$

```python
x = np.array([10,100,1000,10000,100000,1000000,10000000])

(1+(1/x))**x # numpy데이터이기 때문에 이 식을 한번에 계산해준다.
```

어떤 큰 값을 넣어도 2.71828..로 수렴 : 자연상수 e의 탄생

## 로그함수

```python
np.log(np.exp(1))
```

```python
np.e, np.exp(1)
```

```python
def log(x, base):
    return np.log(x) / np.log(base)
```

```python
x1 = np.linspace(0.0001,5,100)
x2 = np.linspace(0.01,5, 100)

y11, y12 = log(x1, 10), log(x2 , np.e)
y21, y22 = log(x1, 1/110), log(x2 , 1/np.e)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x1,y11, label ='$\log_{10} x$', color='k')
ax[0].plot(x2,y12,'--', label ='$\log_{e} x$', color='k')

ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$')
ax[0].legend(fontsize= 15,loc='lower right')

ax[1].plot(x1,y21, label ='$\log_{1/10} x$', color='k')
ax[1].plot(x2,y22,'--', label ='$\log_{1/e} x$', color='k')

ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$')
ax[1].legend(fontsize= 15,loc='upper right')

plt.show()
```

## **시그모이드**

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

- 0과 1사이의 값을 가진다

```python
z = np.linspace(-10,10,100)
sigma = 1/(1+ np.exp(-z))

plt.figure(figsize=(12,8))
plt.plot(z, sigma)
plt.xlabel('$z$',fontsize=15)
plt.ylabel('$\sigma(z)$',fontsize=15)
plt.show()
```

```python
u = np.linspace(0,1,30)
v = np.linspace(0,1,30)

U, V = np.meshgrid(u,v)

z = (1+ U**2) + V / (1 + V**2)
z
```

```python
fig = plt.figure(figsize=(7,7))
ax = plt.axes(projection = '3d')

ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
ax.zaxis.set_tick_params(labelsize=15)

ax.set_xlabel('$x$',fontsize=15)
ax.set_ylabel('$y$',fontsize=15)
ax.set_zlabel('$z$',fontsize=15)

ax.scatter3D(U,V,z,marker='.')
```

함수의 합성 예제

```python
x = np.linspace(-4,4,100)
y = x**3 -15*x + 30 #f(x)
z = np.log(y)       #g(y)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x,y, label='$x^3 - 15x + 30$')
ax[0].legend(fontsize=15)

ax[1].plot(y,z,label='$\log{y}$')
ax[1].legend(fontsize=15)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x,z, label='$\log(f(x))$')
ax[0].legend(fontsize=15)

ax[1].plot(x,y,label='$x^3 - 15x + 30$')
ax[1].legend(fontsize=15)

#한 그래프에 합체
ax_tmp = ax[1].twinx()
ax_tmp.plot(x,z,'--', label='$\log(f(x))$',color='k')
```

log는 x가 크면 조금만 변화하고 작으면 많이 변화한다!
