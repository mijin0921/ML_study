#### 엔트로피 개념
- 확률 분포의 무질서도나 불확실성 혹은 정보 부담 정도를 나타내는 정보 엔트로피 개념을 고안함
- entropy : 얼마만큼의 정보를 담고있는가? 무질서도를 의미, 불확실성을 의미
- 엔트로피 지수가 내려가면 정보의 질서가 잡혀진것이다. 

```python
import numpy as np

p = np.arange(0.001, 1, 0.001)
plt.plot(p, -p * np.log2(p));
plt.grid()
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
```

고유한 값의 빈도수 세기

- np.unique()를 사용하여 고유한 값 얻기: np.unique(array) 함수는 배열의 고유한 값을 반환합니다. 
- np.bincount()를 사용하여 빈도수 계산: np.bincount(array) 함수는 각 고유한 값의 빈도수를 계산합니다

```python
np.bincount(y_pred_tr)
```

```python
#정확도 계산
accuracy_score(iris.target,y_pred_tr)
```

## 새롭게 다시 시작

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

features = iris.data[:,2:]
labels = iris.target
```

```python
# X에는 feature들 y에는 label들을 넣어서 나눈다.
X_train,X_test, y_train,y_test = train_test_split(features,labels,test_size=0.2, random_state=13)
```

```python
X_train.shape, X_test.shape
```

```python
#데이터가 골고루 나눠졌는지 확인해야함. 지금은 골고루 안나눠져 있음. 나쁜건아니지만 일반적으로 고르게 맞춰줌
np.unique(y_test, return_counts=True)
```

```python
# 그럴떄 쓰는 옵션 stratify
X_train,X_test, y_train,y_test = train_test_split(features,labels,
                                                  test_size=0.2, 
                                                  random_state=13,
                                                  stratify=labels)
```

```python
np.unique(y_test, return_counts=True)
```

```python
# 모델의 성능 제한 : max_depth

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12,8))
plot_tree(iris_tree);
```

```python
y_pred_tr = iris_tree.predict(iris.data[:,2:])
accuracy_score(iris.target, y_pred_tr)
```

```python
plt.figure(figsize=(14,8))
plot_decision_regions(X_train, y= y_train, clf=iris_tree, legend=2)
plt.show()
```

훈련용 데이터에 대한 결정 경계가 아까보다 단순화 되었다.

그럼 검증데이터로 테스트 해보자

```python
y_predict = iris_tree.predict(X_test)
accuracy_score(y_test,y_predict)
```

```python
plt.figure(figsize=(14,8))
plot_decision_regions(X_test, y= y_test, clf=iris_tree, legend=2)
plt.show()
```

```python
scatter_highlight_kwargs= {'s':150, 'label': 'Test data', 'alpha': 0.9}
scatter_kwargs = {'s':120, 'edgecolor': None, 'alpha': 0.9}

plt.figure(figsize=(12,8))
plot_decision_regions(features, y = labels, X_highlight=X_test, clf=iris_tree, legend =2 ,
                      scatter_highlight_kwargs=scatter_highlight_kwargs,
                      scatter_kwargs = scatter_kwargs,
                      contourf_kwargs={'alpha':0.2})
```
